**1.1 Study distribution of target variable - Price**

Based on the description above, quick observations about target variable Price: 
- There are NO nan in the target variable
- There appear to be outliers in the target variable   
- Some cars with price 0. Estimating the volume of such cases and decide what to do about them
- Determine if the target variable requires some transformation 

```python
#Show the outliers of car prices using boxplot 
plt.boxplot(data['price']) 
plt.ylabel('Frequency')
plt.title('BoxPlot of Prices') 
plt.show()
```

```python
#Compute percentailes, IQRs etc. to determine the best upper bound and lower bounds to eliminate outliers from the variable price 
price_min = data['price'].min()
price_ntile05 = data['price'].quantile(0.05)
price_ntile10 = data['price'].quantile(0.10)
price_ntile25 = data['price'].quantile(0.25)
price_ntile75 = data['price'].quantile(0.75)
price_ntile90 = data['price'].quantile(0.90)
price_ntile95 = data['price'].quantile(0.95)
price_ntile99 = data['price'].quantile(0.99)
price_max = data['price'].max()
IQR = price_ntile75 - price_ntile25
lower_bound_with_IQR = price_ntile25 - 1.5 * IQR
upper_bound_with_IQR = price_ntile75 + 1.5 * IQR 
volume_with_price0 = len(data[data['price']<=0])/len(data['price'])  

print(f'price_min : {price_min}')
print(f'price_ntile05 : {price_ntile05}')
print(f'price_ntile10 : {price_ntile10}')
print(f'price_ntile25 : {price_ntile25}')
print(f'price_ntile75 : {price_ntile75}')
print(f'price_ntile90 : {price_ntile90}')
print(f'price_ntile95 : {price_ntile95}')
print(f'price_ntile99 : {price_ntile99}')
print(f'price_max : {price_max}')
print(f'lower_bound_with_IQR : {lower_bound_with_IQR}')
print(f'upper_bound_with_IQR : {upper_bound_with_IQR}')
print(f'volume_with_price0 : {volume_with_price0}')
```

**Notes:** Based on the above computation, it appears that roughly 8% prices are 0 and 10 percentile is about 500. I feel that any target price below $500 could be an error. At the same time I am keeping the uppber bound as 75_percentile + 1.5*IQR which is 57,364 

```python
#Filtering out prices below $500 and above 75_percentile+1.5*IQR 
data_filtered_for_price = data[(data['price'] >= 500) & (data['price'] <= upper_bound_with_IQR)]
```

```python
#Show the box plot and histogram of filtered prices 
plt.figure(figsize=(30, 10))

#Show the outliers of car prices using boxplot 
plt.subplot(1, 2, 1)  # 1 row, 2 columns, first subplot 
plt.boxplot(data_filtered_for_price['price']) 
plt.ylabel('Frequency')
plt.title('BoxPlot of Prices') 

# Plotting distribution of prices 
plt.subplot(1, 2, 2)  # 1 row, 2 columns, first subplot 
n, bins, patches = plt.hist(data_filtered_for_price['price'])
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.title('Histogram of Prices') 
for i in range(len(n)):
    plt.text(bins[i] + (bins[1] - bins[0]) / 2, n[i], str(int(n[i])), ha='center', va='bottom')
```

**Notes:** The filtered data continues to be right skewed. I will perform log, squareroot and box-cox transformation and determine which of these I should be using for modeling 

```python
#Show the distributions using log and squateroot transformation 
data_org = data_filtered_for_price['price']
data_log = np.log(data_org)  
data_sqrt = np.sqrt(data_org)

# Create a DataFrame for easy plotting
df = pd.DataFrame({
    'Original': data_org,
    'Log Transformed': data_log,
    'Square Root Transformed': data_sqrt
})

# Plot distributions
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
for ax, col in zip(axes, df.columns):
    sns.histplot(df[col], kde=True, ax=ax, bins=30, color='skyblue')
    ax.set_title(f'{col} Distribution')
    ax.set_xlabel(col)
    ax.set_ylabel('Frequency')
plt.tight_layout() 
plt.show()
```

**Notes:** I like how square root transformation is leading to a normal distribution. I will transform the target variable price using squareroot transformation 

```python
#Transforming target variable Price with square root transformation 
data['price_sqrt'] = np.where((data['price'] >= 500) & (data['price'] <= upper_bound_with_IQR),np.sqrt(data['price']),np.nan)
```

```python
#Displaying final ditribution of target variable price_sqrt 
sns.histplot(data['price_sqrt'], kde=True, bins=30, color='skyblue')
plt.title('Distribution of Price of Car (sqrt)')
plt.xlabel('Price_SQRT')
plt.ylabel('Frequency')
plt.show()
```

```python
#Splitting data into training and test datasets by split indices 
data_clean_encoded.reset_index(drop=True)
indices = data_clean_encoded.index
train_idx, dev_idx = train_test_split(indices, test_size=0.3, random_state=42)
data_clean_encoded_train = data_clean_encoded.loc[train_idx]
data_clean_encoded_dev = data_clean_encoded.loc[dev_idx]

print(data_clean_encoded_train.shape)
print(data_clean_encoded_dev.shape)
```

```python
#Target Encoding variables - manufacturer, state 

#computing target mean using training dataset only 
manufacturer_target_mean = data_clean_encoded_train.groupby('manufacturer')['price_sqrt'].mean()
state_target_mean = data_clean_encoded_train.groupby('state')['price_sqrt'].mean()

#mapping back to entire dataset using target means 
data_clean_encoded['manufacturer_encoded'] = data_clean_encoded['manufacturer'].map(manufacturer_target_mean)
data_clean_encoded['state_encoded'] = data_clean_encoded['state'].map(state_target_mean)

data_clean_encoded = data_clean_encoded.drop(columns=['manufacturer','state'])
data_clean_encoded_train = data_clean_encoded.loc[train_idx]
data_clean_encoded_dev = data_clean_encoded.loc[dev_idx]
```

**Model 6: Building Ridge regression with Grid Search CV to find optimal alpha**

```python
#Build model 

y = data_clean_encoded['price_sqrt'] 
X = data_clean_encoded.drop('price_sqrt', axis = 1)

#indices = data_clean_encoded.index
#train_idx, dev_idx = train_test_split(indices, test_size=0.3, random_state=42)

pipe = Pipeline([('poly_features', PolynomialFeatures(degree = 2, include_bias = False)), 
                 ('scale',StandardScaler()),
                 ('ridge', Ridge())])

alphas_to_try = {'ridge__alpha': 10**np.linspace(-5,4,100)}

model_finder = GridSearchCV(estimator = pipe,
                           param_grid = alphas_to_try,
                           scoring = "neg_mean_squared_error",
                           cv=5)

model_finder.fit(X_train, y_train) 
```

```python
best_model = model_finder.best_estimator_
print(best_model)
```

**Notes:** GridSearchCV has found best alpha to be 43.29

```python
#Predict values and compute error 
train_predict = model_finder.best_estimator_.predict(X_train)
dev_predict = model_finder.best_estimator_.predict(X_dev) 

train_mse = mean_squared_error(y_train, train_predict)
train_r2 = r2_score(y_train, train_predict)
dev_mse = mean_squared_error(y_dev, dev_predict)
dev_r2 = r2_score(y_dev, dev_predict)

model_result = pd.DataFrame({
                              'Model': 'Ridge Reg w/ GridSearched Alpha',
                              'MSE_Train': [int(train_mse)],
                              'MSE_Dev': [int(dev_mse)],
                              'R2_Train': [round(train_r2,2)],
                              'R2_Dev': [round(dev_r2,2)]
                          })

model_results = pd.concat([model_results, model_result], ignore_index=True)
print(model_results)
```

```python
# Get the coefficients of the model 
feature_names = model_finder.best_estimator_.named_steps['poly_features'].get_feature_names_out()
ridge_coefs = model_finder.best_estimator_.named_steps['ridge'].coef_
ridge_df = pd.DataFrame({'feature':feature_names, 'coef':ridge_coefs, 'coef_abs':abs(ridge_coefs)})
```

**Model 7: Building Lasso regression after fitting LassoCV to find best alpha** 

```python
#Build model 
pipe = Pipeline([('poly_features', PolynomialFeatures(degree = 2, include_bias = False)), 
                 ('scale',StandardScaler()),
                 ('lassoCV', LassoCV(cv=5, max_iter=5000, random_state=42))])
pipe.fit(X_train, y_train)
```

```python
#Finding the best alpha using LassoCV 
best_alpha = pipe.named_steps['lassoCV'].alpha_
print(best_alpha)
```

```python
#Predict values and compute error 
train_predict = final_pipe.predict(X_train)
dev_predict = final_pipe.predict(X_dev) 

train_mse = mean_squared_error(y_train, train_predict)
train_r2 = r2_score(y_train, train_predict)
dev_mse = mean_squared_error(y_dev, dev_predict)
dev_r2 = r2_score(y_dev, dev_predict)

model_result = pd.DataFrame({
                              'Model': 'Lasso Regression w/ optimal alpha',
                              'MSE_Train': [int(train_mse)],
                              'MSE_Dev': [int(dev_mse)],
                              'R2_Train': [round(train_r2,2)],
                              'R2_Dev': [round(dev_r2,2)]
                          })

model_results = pd.concat([model_results, model_result], ignore_index=True)
```

```python
# Get the coefficients of the model 
lasso_feature_names = final_pipe.named_steps['poly_features'].get_feature_names_out()
lasso_coefs = final_pipe.named_steps['lasso'].coef_
lasso_df = pd.DataFrame({'feature':lasso_feature_names, 'coef':lasso_coefs, 'coef_abs':abs(lasso_coefs)})
```
